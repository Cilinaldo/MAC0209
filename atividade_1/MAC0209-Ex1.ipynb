```python
# Com o imageio podemos carregar uma imagem que está na mesma pasta que o jupyter notebook
img_exemplo = imageio.imread('minhafoto.jpeg')
```

```python
# Para exibir a imagem carregada primeiro criamos uma 'figura'
fig, ax = plt.subplots(1)

# Aqui removemos os eixos horizontal e vertical (x e y)
# Descomente a linha abaixo para remover os eixos horizontal e vertical
#ax.axis('off')

# Aqui preenchemos a 'figura' com a imagem que carregamos com o imageio
ax.imshow(img_exemplo)
```

```python
# Ou também podemos usar diretamente o comando "plt.imshow"

plt.imshow(img_exemplo)
```

Note que a imagem está representada por uma matriz. No caso de uma imagem colorida temos um conjunto de 3 matrizes concatenadas, uma representando a intensidade de vermelho, uma a de verde a por último a intensidade da cor azul (i.e. RGB).

```python
# Aqui inspecionamos as primeiras 10 colunas da primeira linha do canal de cor azul

# Note que o primeiro índice indica a coluna, o segundo a linha e o terceiro o canal de cor (RGB)
img_exemplo[:10, 0, 2]
```

Curiosidade: podemos realizar operações como refletir ao redor de um eixo:

```python
# Aqui invertemos a matriz (imagem) ao longo do eixo horizontal
inv_x = img_exemplo[:, ::-1, :]

# Aqui ao longo do eixo vertical
inv_y = img_exemplo[::-1, :, :]

# Agora vamos exibir duas imagens na mesma figura
fig, ax = plt.subplots(1, 2)

ax[0].imshow(inv_x)
ax[1].imshow(inv_y)
```
